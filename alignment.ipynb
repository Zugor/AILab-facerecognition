```python
for imgFile in imgFiles:
    try:
        img = cv2.imread('ImageCall/'+imgFile)
        rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        h, w, c = rgb.shape
        if w > h:
            rotated = imutils.rotate_bound(rgb, -90)
            img = cv2.cvtColor(rotated, cv2.COLOR_RGB2BGR)
        cv2.imwrite('ImageFinal/'+imgFile,img)
    except:
        print(imgFile)
```
